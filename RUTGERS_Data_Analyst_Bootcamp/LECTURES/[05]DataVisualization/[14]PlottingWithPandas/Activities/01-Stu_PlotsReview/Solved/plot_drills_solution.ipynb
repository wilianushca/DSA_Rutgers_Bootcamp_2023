```python
# Import Dependencies
# Importing Usual Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA SET 1
# two datasets of data in lists [" 1 "," 2 "," 3 "]
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
# Used < np.arrange(start,end) > with length of gyms (4)
x_axis = np.arange(0, len(gyms))
# Set a empty list []
    # For loop for each 'x_axis' to < .append( ) > add each x in tick_locations
tick_locations = []
for x in x_axis:
    tick_locations.append(x)
# < plt.title( ) >
plt.title("NYC Gym Popularity")
# < plt.xlabel( ) > for the x axis
plt.xlabel("Gym Name")
# < plt.ylabel( ) > for the y axis
plt.ylabel("Number of Members")
# < plt.xlim(start, end) >
plt.xlim(-0.75, len(gyms)-.25)
# < plt.ylim(start,end) >
plt.ylim(0, max(members) + 5)
# < plt.bar(x,y,facecolor=" ", alpha=, align="center" ) >
### facecolor = bar color
### alpha= transparency
plt.bar(x_axis, members, facecolor="red", alpha=0.75, align="center")
# < plt.xticks(ticklocation,ticknames) >
plt.xticks(tick_locations, gyms) 
plt.show()
```

```python
# DATA SET 2
# one data axis is 2 times pi
# second data axis is a list of 0 through 2 times pi counting by 0.1
# then we use < np.sin > to find the sin of each value
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
# Used < plt.title() >
plt.title("Sin from 0 to 2$\pi$")
# Used < plt.xlabel() >
plt.xlabel("Real Numbers from 0 to 2$\pi$")
# Used < plt.ylabel() >
plt.ylabel("sin(x)")
# Used < plt.hlines(x,y,how long horizontaly,transparency) >
plt.hlines(0, 0, x_lim, alpha=0.2)
# Used < plt.xlim(start,end) > for x axis limits
plt.xlim(0, x_lim)
# Used < plt.ylim(start,end) > for y axis limits
plt.ylim(-1.25, 1.25)
# Used < plt.plot(x,y,circle marker,red line,size of line=1) >
plt.plot(x_axis, sin, marker="o", color="red", linewidth=1)
plt.show()
```

```python
# DATA SET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
# Used < plt.title() >
plt.title("NYC Gym Popularity")
# Used < plt.pie(data, explode= popout,labels=,colors=, autopct=, shadow==, startangle=) >
# explode= popout
# labels= labels
# colors= colors
# autopct= the format % f%
# shadow= shadow
# startangle= angle counterclockwise to start the angle
plt.pie(members, explode=explode, labels=gyms, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
# < plt.axis() > and set to equal to make a full circle
plt.axis("equal")

plt.show()
```

```python
# DATA SET 4
# <np.arrange(start,end,countby)>
x_axis = np.arange(0, 10, 0.1)
times = []
# for loop in 'x_axis' to add times to the times []
# <np.random.randint(start,end)>
# <np.ceil(max())> meaning celing or highest datapoint
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
# <plt.title()>
plt.title("Running Time of FakeSort for Sample Input Sizes")
# <plt.xlabel()> to title the x-axis
plt.xlabel("Length of Input Array")
# <plt.ylabel()> to title the y-axis
plt.ylabel("Time to Sort (s)")
# <plt.scatter(x,y,circle marker,red color)>
plt.scatter(x_axis, times, marker="o", color="red")
# Displaying data
plt.show()
```
