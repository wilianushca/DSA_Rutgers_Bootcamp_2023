```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example outlier plot of reaction times
# <[]> to insert a list of times
times = [96,98,100,105,85,88,95,100,101,102,97,98,5]
# <fig1, ax1> meaning we created a figure and an axis
# <plt.subplots()> using this we can create the figure and axis
fig1, ax1 = plt.subplots()
# <.set_title()> to set the title to 'Reaction Times at Baseball Batting Cage'
ax1.set_title('Reaction Times at Baseball Batting Cage')
# <.set_ylabel()> to set the y-axis label to 'Reaction Time (ms)'
ax1.set_ylabel('Reaction Time (ms)')
# <.boxplot()> to set a boxplot from 'times'
ax1.boxplot(times)
# <plt.show()> to display data
plt.show()
```

```python
# We need to sort the data to determine which could be outliers
# <.sort()> to sort the times from least to greatest
times.sort()
# Display "times"
print(times)
```

```python
# The second example again looks at the LAX temperature data set and computes quantiles
# <pd.read_csv('PATH')> to read the csv doc.
temperature_df = pd.read_csv('../Resources/lax_temperature.csv')
# <[]> Selected a list ['HourlyDryBulbTemperature']
temperatures = temperature_df['HourlyDryBulbTemperature']

# <fig1, ax1> meaning we created a figure and an axis
# <plt.subplots()> using this we can create the figure and axis
fig1, ax1 = plt.subplots()
# <.set_title()> to set the title to 'Temperatures at LAX'
ax1.set_title('Temperatures at LAX')
# <.set_ylabel()> to set the y-axis label to 'Temperature (°F)'
ax1.set_ylabel('Temperature (°F)')
# <.boxplot()> to set a boxplot from 'temperatures'
ax1.boxplot(temperatures)
# <plt.show()> to display data
plt.show()
```

```python
# If the data is in a dataframe, we use pandas to give quartile calculations
# <.quantile([.25,.5,.75])> to figure out the first second and third quantile
quartiles = temperatures.quantile([.25,.5,.75])
# first quartile = 0.25
lowerq = quartiles[0.25]
# third quartile = 0.75
upperq = quartiles[0.75]
# interquartile range is upper minus lower
iqr = upperq-lowerq

print(f"The lower quartile of temperatures is: {lowerq}")
print(f"The upper quartile of temperatures is: {upperq}")
print(f"The interquartile range of temperatures is: {iqr}")
print(f"The the median of temperatures is: {quartiles[0.5]} ")

# Boundaries using IQR Formula +- (1.5*iqr)
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```
