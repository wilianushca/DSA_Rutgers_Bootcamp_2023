```python
# Imports Usual Dependencies
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Creating two lists {[temp],[sales]} to add data points

temp = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
sales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
```

```python
# Tell matplotlib to create a scatter plot based upon the above data

# Without scoop_price
#plt.scatter(temp, sales, marker="o", facecolors="red", edgecolors="black")

# BONUS: With scoop_price set to the scalar value
    # < [ ] > 
    #>>> List of scoop prices [89, 18, 10, etc. ]
        # < plt.scatter('x-axis', 'y_axis', marker=" ", facecolors=" ", edgecolors=" ", s='Scale' >
        #>>> Plotted with ['temp','sales'] as axis', with circle markers, red circles, black outlines, 'scoop_price' as list of scales
            # < plt.show( ) >
            #>>> Displays the data

scoop_price = [89, 18, 10, 28, 79, 46, 29, 38, 89, 26, 45, 62]
plt.scatter(temp, sales, marker="o", facecolors="red", edgecolors="black", s=scoop_price)
plt.show()
```

```python
# Set the upper and lower limits of our y axis
    # < plt.ylim( start, end ) >
    #>>> Limits the plot from 180 to 620
        # < plt.show( ) >
        #>>> Displays data

plt.ylim(180,620)
plt.show()
```

```python
# Set the upper and lower limits of our x axis
    # < plt.xlim( start, end ) >
    #>>> Sets limit on x-axis from 11 to 26
        # < plt.show( ) >
        #>>> Display the data

plt.xlim(11,26)
plt.show()
```

```python
# Create a title, x label, and y label for our chart
    # < plt.title( ) >
    #>>> Sets title to "Ice Cream Sales v Temperature"
        # < plt.xlabel( ) >
        #>>> Sets x-axis title to "Temperature(Celsius)"
            # < plt.ylabel ( ) >
            #>>> Sets y-axis title to "Sales (Dollars)"
                # < plt.show( ) >
                #>>> Displays data
                
plt.title("Ice Cream Sales v Temperature")
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Sales (Dollars)")
plt.show()
```

```python
# Save an image of the chart and print to screen
    # < plt.savefig(PATH) >
    #>>> Saves image to the desired path
        # < plt.show( ) >
        #>>> Displays data

plt.savefig("../Images/IceCreamSales.png")
plt.show()
```
