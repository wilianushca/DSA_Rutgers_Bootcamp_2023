```python
# Import Dependencies
    # Import usual dependencies

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the X and Y axis lists
    # Used < np.arrange( start, end, countby ) > to make 'time'
        # Used [ ] to make a list for 'danger_drop_speeds'
            # Used [ ] to make a list for 'railgun_speeds'

time = np.arange(0,130,10)
danger_drop_speeds = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun_speeds = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
# Plot the charts and apply some styling
    # Used < plt.plot ( line1, line2, color=, label= ) > to make a plot named 'danger_drop' 
        # Used < plt.plot( line1, line2, color=, label= ) > to make a plot named 'railgun' 
        # NOTE: Lines have to have the [,] at the end... ex.( danger_drop[,] & railgun[,] )

danger_drop, = plt.plot(time, danger_drop_speeds, color="red", label="Danger Drop")
railgun, = plt.plot(time, railgun_speeds, color="blue", label="RailGun")
plt.show()
```

```python
# Add labels to X and Y axes :: Add title
    # Used < plt.title( ) > to title it "Coaster Speed Over Time"
        # Used < plt.xlabel( ) > to title x-axis to "Coaster Runtime"
            # Used < plt.ylabel( ) > to title y-axis to "Speed(MPH)"
                # Used < plt.show( ) > to display the dataplot 

plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
plt.show()
```

```python
# Set the limits for the X and Y axes
    # Used < plt.xlim( start, end ) > to set x-limits to [0:120]
        # Used < plt.ylim( start, end ) > to set y-limits to [5:95]
            # Used < plt.show( ) > to display the data

plt.xlim(0,120)
plt.ylim(5,95)
plt.show()
```

```python
# Create a legend for the chart
    # Used < plt.legend( handles= [line1, line2], loc=" " ) > to make a legend of both lines in the best location 
        # Used < plt.show( ) >

plt.legend(handles=[danger_drop, railgun], loc="best")
 
```

```python
# Add in a grid for the chart
    # Used < plt.grid( ) > to add grid lines to the plot 
        # Used < plt.show ( ) > to display the plot table

plt.grid()
plt.show()
```
