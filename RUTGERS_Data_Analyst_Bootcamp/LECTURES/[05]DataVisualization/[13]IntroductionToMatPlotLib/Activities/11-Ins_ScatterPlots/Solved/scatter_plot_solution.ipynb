```python
# Import Dependencies

import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The maximum x value for our chart will be 100

x_limit = 100

# List of values from 0 to 100 each value being 1 greater than the last
    # 'x_limit' == 100
    # < np.arrange( start, end, countby ) > == inserting a list [ ] from 0-100, counting by 1

x_axis = np.arange(0, x_limit, 1)

# Create a random array of data that we will use for our y values
    # < random.random( ) for ' ' in ' '] > 
    # >>> list comprehension with 'x_axis' to pick 100 random points from 0 to 1 and naming it 'data'

data = [random.random() for value in x_axis]
```

```python
# Tells matplotlib that we want to make a scatter plot
# The size of each point on our plot is determined by their x value
    # < plt.scatter( x_data, y_data, marker='o', label='Data Points' ) > 
    #>>> Inserted x, y axis' [ 'x_axis', 'data' ], circle marker, red circles, black outlines, size for the markers, and transparency of 0.75
        # < plt.show( ) > 
        #>>> Displays data

plt.scatter(x_axis, data, marker="o", facecolors="red", edgecolors="black",
            s=x_axis, alpha=0.75)
plt.show()
```

```python
# The y limits of our scatter plot is 0 to 1
    # < plt.ylim( start, end ) >
    #>>> Sets a y-axis limit from 0 to 1
        # < plt.show( ) >
        #>>> Displays data

plt.ylim(0, 1)
plt.show()
```

```python
# The x limits of our scatter plot is 0 to 100
    # < plt.xlim( start, end ) >
    #>>> Limits the x-axis from 0 to 100
        # < plt.show( ) >
        #>>> Displays data

plt.xlim(0, x_limit)
plt.show()
```
