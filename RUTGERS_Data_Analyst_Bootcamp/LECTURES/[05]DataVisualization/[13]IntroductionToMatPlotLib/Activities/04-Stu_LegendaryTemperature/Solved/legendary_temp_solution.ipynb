```python
# Dependencies
    # Importing usual Dependencies 

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis to numerical value for month
    # Use < np.arrange( start, end, countby ) > to set the 'x_axis' 
    # Displayed 'x_axis'

x_axis = np.arange(1,13,1)
x_axis
```

```python
# Avearge weather temp
    # Inserted a list [n1,n2,n3,etc.] to set up 'points_f'

points_F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert to Celsius C = (F-32) * 0.56
    # Used a for loop < x in 'points_f' > to subtract each 'x'by 32 and then multiply by 0.56 
        # Variable name is 'points_c'

points_C = [(x-32) * 0.56 for x in points_F]
points_C
```

```python
# Create a handle for each plot
    # Created a plot < plt.plot( ) > with 'x_axis','points_F' with a "+" [plus] marker and "blue" color with a linewdidth of 1 and the line will be called "Fahrenheit"
        # Created a plot < plt.plot( ) > with 'x_axis','points_C' with a "S" [square] marker and "red" color with a linewdidth of 1 and the line will be called "Celsius"
            # Used < plt.show( ) > to display

fahrenheit, = plt.plot(x_axis, points_F, marker="+",color="blue", linewidth=1, label="Fahrenheit")
celsius, = plt.plot(x_axis, points_C, marker="s", color="Red", linewidth=1, label="Celsius")
plt.show()
```

```python
# Set our legend to where the chart thinks is best
    # Used < plt.legend( handles=[ line1, line2, etc. ], loc="best" ) > to show the legend
        # < plt.show( ) > to display

plt.legend(handles=[fahrenheit, celsius], loc="best")
plt.show()
```

```python
# Create labels for the X and Y axis
    # Used < plt.xlabel( ) > to label it "Months"
        # Used < plt.ylabel( ) > to label it "Degrees"
            # Used < plt.show( ) > to display the plot

plt.xlabel("Months")
plt.ylabel("Degrees")
plt.show()
```

```python
# Save and display the chart
    # Used < plt.savefig( ) > to save into the "PATH" and save an image
        # Used < plt.show( ) > to display the data

plt.savefig("../Images/avg_temp.png")
plt.show()
```
