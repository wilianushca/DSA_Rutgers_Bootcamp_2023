```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv

    # < pd.read_csv(PATH) >
    #>>> Reads from 'avg_rain_state.csv' & turns to a panda

        # < rain_df.head() >
        #>>> Displays the 'rain_df' data 
        
rain_df = pd.read_csv("../Resources/avg_rain_state.csv")
rain_df.head()
```

```python
# Set x axis and tick locations

    # < np.arange(len(' ')) >
    #>>> Creates a range of data from the length of points in 'rain_df' (49 points)

        # < [value+0.4 for ' ' in ' '] >
        #>>> List comprehension [ ] and adding a 0.4 in every value in 'x-axis'

x_axis = np.arange(len(rain_df))
tick_locations = [value+0.4 for value in x_axis]
tick_locations
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space

    # < plt.figure(figsize=( width, height )) >
    #>>> Chose the width (20 inches) and height (4 inches)

        # < plt.bar(x_axis, y-axis, color=' ', alpha=' ', align=" ") >
        #>>> Align=" " == A parameter specifying alignment of the bars. It can take values like 'center', 'edge',
        #>>> Used ['x-axis','rain_df["Inches"]'] as x/y axis', with red bars and a transparency of 0.5 while bars are aligned by edge

            # < plt.xticks( tick_positions, tick_labels, rotation=" ") >
            #>>> Adds tick markers from 'tick_locations', labels from 'rain_df["State"]' column and a rotation of the labels to be vertical

                # < plt.show( ) >
                #>>> Displays the data

plt.figure(figsize=(20,4))
plt.bar(x_axis, rain_df["Inches"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, rain_df["State"], rotation="vertical")
plt.show()
```

```python
# Set x and y limits

    # < plt.xlim( start, end ) >
    #>>> Set the minimum at -0.25 and maximum at the length of 'x_axis'

        # < plt.ylim( start, end ) >
        #>>> Set the minimum at 0 and maximum at the highest data point in the column "Inches" from 'rain_df' and adding 10 

            # < plt.show( ) >
            #>>> Displays data

plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(rain_df["Inches"])+10)
plt.show()
```

```python
# Set a Title and labels

    # < plt.title( ) >
    #>>> Sets the title to "Average Rain per State"

        # < plt.xlabel( ) >
        #>>> Sets the x-axis label to "State"

            # < plt.ylabel( ) >
            #>>> Sets the y-axis label to "Average Amount of Rainfall in Inches"

                # < plt.show( ) >
                #>>> Displays the data

plt.title("Average Rain per State")
plt.xlabel("State")
plt.ylabel("Average Amount of Rainfall in Inches")
plt.show()
```

```python
# Save our graph and show the graph

    # < plt.tight_layout( ) >
    #>>> Automatically adjust the subplot parameters to ensure that the contents of the plot fit within the figure area without overlapping

        # < plt.savefig( ) >
        #>>> Saves an image into the directed "Path"

            # < plt.show( ) >
            #>>> Displays the data

plt.tight_layout()
plt.savefig("../Images/avg_state_rain.png")
plt.show()
```
